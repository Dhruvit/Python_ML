```
from sklearn import datasets
from sklearn import metrics # to masure how model it is
from sklearn.svm import SVC # support vector classifier, support vector machine
```

```
ds = datasets.load_iris() #import data
```

```
model = SVC() #support vector classifier, use the model
model.fit(ds.data, ds.target) #fit data into model
#print(ds.data)
#print(ds.target)
print(model) #additional info about model
```

```
expected = ds.target
predicted = model.predict(ds.data)
print(predicted)
print(metrics.classification_report(expected, predicted))
# 99% accuracy
```
